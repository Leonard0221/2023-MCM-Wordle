```python
# Using readlines()
#wordle_answer.txt
# file1 = open('wordle_answer.txt', 'r')
file1 = open('all_five_letter.txt', 'r')
Lines = file1.readlines()
  
count = 0
# Strips the newline character‘
word_list = []
for line in Lines:
    count += 1
    word_list.append(line.strip())
file1.close()
```

```python
len(word_list)
```

```python
import nltk
from tqdm import tqdm
word = word_list[0]
l = []
def count_l(l):
    dic = {}
    for i in l:
        if i not in dic:
            dic[i] = 1
        else:
            dic[i] += 1
    return dic

result = []
for word in tqdm(word_list):
    l = []
    for i in word_list[:]:
        l.append(nltk.edit_distance(word, i))
    result.append(count_l(l))
print(result)
```

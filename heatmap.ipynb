```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('Problem_C_Data_Wordle.xlsx', skiprows=1, index_col='Date_c')
df
```

```python
word_array = df['Word'].tolist()
word_array
```

```python
counts = {}

# Iterate over the words and count the alphabets at each index
for word in word_array:
    for i, letter in enumerate(word):
        if letter == " ":
            print(word)
        if i not in counts:
            counts[i] = {}
        if letter not in counts[i]:
            counts[i][letter] = 0
        counts[i][letter] += 1
counts
```

```python
result = pd.DataFrame(counts)
result
```

```python
sns.heatmap(result, cmap='YlGnBu')
```

```python
result.to_csv('output2.csv', index=True)
```

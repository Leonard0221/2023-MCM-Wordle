```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.DataFrame({'Features':['Freq_HAL','I_Zscore','I_SD','short_2','I_Mean_RT','short_4','short_3','all_1','all_3','short_5','all_4','brown_freq','all_2','I_Mean_Accuracy','word_format','Obs','Syllable','short_1','p','all_5'],'Feature Importance': [42,40,37,33,29,29,28,26,25,23,22,19,18,15,12,11,7,7,7,6]})
df.head()
```

```python
sns.set(rc={'figure.figsize':(10,12)})
ax = sns.barplot(y='Features',x='Feature Importance', data=df,palette='GnBu_d')
initialx=0
for p in ax.patches:
  ax.text(p.get_width(),initialx+p.get_height()/8,'{:1.0f}'.format(p.get_width()))
  initialx+=1
```
